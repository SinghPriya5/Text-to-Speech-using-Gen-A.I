```python
from transformers import pipeline
```

```python
text="Priya is a brave and strong-minded tech enthusiast with hands-on experience in Generative AI, dedicated to building intelligent solutions that make an impact."
pipe=pipeline("text-to-speech",model="suno/bark-small",device="cpu")
output=pipe(text)
```

```python
output
```

```python
from IPython.disply import Audio

Audio(output['audio'],rate=output['sampling_rate'])
```
